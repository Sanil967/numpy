```python
x=Data[['sepal.length','sepal.width','petal.length','petal.width']]
x
y=Data[['variety']]
x,y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
x_train,x_test,y_train,y_test
```

```python
reg=LogisticRegression()
reg.fit(x_train,y_train)
```

```python
log=reg.predict(x_test)
log
```

```python
log_a=accuracy_score(y_test,log)
log_a
```

```python
a=confusion_matrix(y_test,log)
a
```

```python
tr=DecisionTreeClassifier()
tr.fit(x_train,y_train)
```

```python
trr=tr.predict(x_test)
trr
```

```python
ssv=SVC(kernel='linear') #i can be poly sigmoid
ssv.fit(x_train,y_train)
```

```python
SVVC=ssv.predict(x_test)
SVVC
```

```python
neigh=KNeighborsClassifier(n_neighbors=3)#it should be odd mostly
neigh.fit(x_train,y_train)
```

```python
nn=neigh.predict(x_test)
nn
```

```python
trr_Acc=accuracy_score(y_test,trr)
trr_Acc
```

```python
SVVC_acc=accuracy_score(y_test,SVVC)
SVVC_acc
```

```python
nn_acc=accuracy_score(y_test,nn)
nn_acc
```

```python
a=[nn_acc,SVVC_acc,trr_Acc,log_a]
b=['KNeighborsClassifier','SVC','DecisionTreeClassifier','LogisticRegression']
plt.bar(b,a)
```

Cluster

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
x=Data[['sepal.length','sepal.width','petal.length','petal.width']]
y=Data[['variety']]
```

```python
agg=AgglomerativeClustering(n_clusters=3)
```

```python
agg.fit(x)
```

```python
label=agg.labels_
label
```

```python
plt.scatter(x.iloc[:,0],x.iloc[:,1],c=label)
```

```python
from sklearn.cluster import KMeans
```

```python
Kmena=KMeans(n_clusters=3)
kkm=Kmena.fit_predict(x)
kkm
```

```python
Data['cluster']=kkm
```

```python
d1=Data[Data.cluster==0]
d2=Data[Data.cluster==1]
d3=Data[Data.cluster==2]
plt.scatter(d1['sepal.length'],d1['sepal.width'],color="red")
plt.scatter(d2['sepal.length'],d2['sepal.width'],color="blue")
plt.scatter(d3['sepal.length'],d3['sepal.width'],color="pink")
plt.scatter(Kmena.cluster_centers_[:,0],Kmena.cluster_centers_[:,1],color='black',marker="*",label='centroid')
```
