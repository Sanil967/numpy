```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1=np.array([12,10,21,11])
x2=np.array([12,22,29,41])
x=np.array([20,10,3,5])
y=np.array([10,20,9,40])
```

```python
sizes=np.array([10,20,100,50])#jitni value array me ha utna hi value isme bhi hona chiaye like array size 4 then size should also have 4 value
plt.scatter(x1,x2,color='red')
plt.scatter(x,y,color='green',s=sizes,alpha=0.5)#alpha value should be between 0 to 1
plt.show()
```

```python
plt.bar(x1,x2)
plt.show()
```

```python
plt.barh(x1,x2)#horizontal bar
```

```python
plt.bar(x1,x2,color='green',width=2)
plt.show()
plt.bar(x,y,color='m',width=1)
plt.show()
```

```python
plt.hist(x)#only single variable
```

```python
plt.pie(x)
```

```python
plt.pie(x)
plt.pie(y)
plt.show()
plt.pie(x1)
plt.pie(x2)
plt.show()
```

```python
std=np.array([20,30,40,10])
std_ame=['apple','mango','kiwi','avacado']
plt.pie(std,labels=std_ame)
```

```python
std=np.array([20,30,40,10])
std_ame=['apple','mango','kiwi','avacado']
plt.pie(std,labels=std_ame,startangle=90)
```

```python
std=np.array([20,30,40,10])
std_ame=['apple','mango','kiwi','avacado']
plt.pie(std,labels=std_ame,startangle=90,explode=[0.1,0,0,0])#with explode jitne array ha utne value
```

```python
std=np.array([20,30,40,10])
std_ame=['apple','mango','kiwi','avacado']
plt.pie(std,labels=std_ame,startangle=90,explode=[0.1,0,0,0],shadow=True)#in explode 
```

```python
std=np.array([20,30,40,10])
std_ame=['apple','mango','kiwi','avacado']
plt.pie(std,labels=std_ame,startangle=90,explode=[0.1,0,0,0],shadow=True)
#plt.legend()#it will give box popout of information 
plt.legend(title="fruit")
plt.show()
```
