```python
import  numpy as np
import pandas as pd
import matplotlib.pyplot as plt#plt is an identifier and any name be used and _ or digit dont put 
```

```python
a1=np.array([[1,2,3,4],[5,2,6,7]])
a2=np.array([[1,2,3,4],[5,2,6,7]])
print(a1)
```

```python
x=np.array([0,20]) # x value
y=np.array([0,250])# y value

plt.plot(x,y)#plot is function that will plot x any y axis on graph
plt.show()#with show it will display garph in proper way without error message
```

```python
plt.plot(x,y,'o')#specify type of marker
plt.show()
```

```python
plt.plot(x,y,'x')#x is  type of marker which will show on start and end point
plt.show()
```

```python
a1=np.array([1,2,3,4])
a2=np.array([5,2,6,7])
plt.plot(a1,a2)
plt.show()
```

```python
plt.plot(a2,marker='o')#
plt.show()
```

```python
plt.plot(a2,marker='*')#star
plt.show()
```

```python
plt.plot(a2,marker='.')#point
plt.show()
```

```python
plt.plot(a2,marker=',')#pixel 
plt.show()
```

```python
plt.plot(a2,marker='x')#x
plt.show()
```

```python
plt.plot(a2,marker='X')#x(filled)
plt.show()
```

```python
plt.plot(a2,marker='P')#plus 
plt.show()
```

```python
plt.plot(a2,marker='D')#Diamond 
plt.show()
```

```python
plt.plot(a2,marker='v')#triangle 
plt.show()
```

```python
plt.plot(a2,marker='|')
plt.show()
```

```python
plt.plot(a2,marker='_')#plus 
plt.show()
```

```python
plt.plot(a2,marker='*',ms=10,mec='r',mfc='y')#ms=marker size  ,mec=marker edge color, mfc=marker face color
plt.show()
```

```python
plt.plot(a2,marker='*',ms=10,mec='r',mfc='y',ls=":")#ls=line style
plt.show()
```

```python
plt.plot(a2,marker='*',ms=10,mec='r',mfc='y',ls="-")#ls=line style
plt.show()
```

```python
plt.plot(a1,a2,linestyle=':',color='g',linewidth='5')#ls=line style
plt.show()
```

```python
plt.plot(x,y,a1,a2)
plt.show()
```

```python
font={'color':'green','size':'20'}
plt.title("line chart",fontdict=font,loc="left")#without loc text will come in center else left
plt.xlabel("no1")
plt.ylabel("no2")
plt.plot(x,y)
plt.grid(axis='x',color='red',linestyle='--',linewidth='0.5')
```

```python
plt.subplot(1,2,1)#1 is row # 2 is graph
plt.title("p1")
plt.plot(x,y)
plt.subplot(1,2,2)
plt.title("p2")
plt.plot(a1,a2)
```
