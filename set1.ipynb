```python
x=aa[["sepal.length","sepal.width","petal.length","petal.width"]]
y=aa['variety']
x,y
```

```python
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.2)
x_test,x_train,y_test,y_train
```

```python
reg=LogisticRegression()
reg.fit(x_train,y_train)
```

```python
log_prd=reg.predict(x_test)
log_prd
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
acc=accuracy_score(y_test,log_prd)
acc
```

```python
con=confusion_matrix(y_test,log_prd)
con
```

```python
from sklearn.svm import SVC
svmaa=SVC(kernel='linear')
```

```python
svmaa.fit(x_train,y_train)
```

```python
svm_pred=svmaa.predict(x_test)#kernel can be 'linear','poly','sigmoid','rbf'
svm_pred
```

```python
aad=accuracy_score(y_test,svm_pred)
aad
```

```python
from sklearn.tree import DecisionTreeClassifier
dec_reg=DecisionTreeClassifier()
dec_reg.fit(x_train,y_train)
```

```python
dec=dec_reg.predict(x_test)
dec
```

```python
accd=accuracy_score(y_test,dec)
accd
```

```python
cmc=confusion_matrix(y_test,dec)
cmc
```

```python
arr=[aad,accd]
label=("svm","decisontree")
plt.bar(label,arr)
```
